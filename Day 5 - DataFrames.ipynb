### Merge Data 

```python
movies = pd.DataFrame({'movieid':[1,2,3,4,5],
                       'title':['Toy Story 2','Mission Impossible','Avatar','Blade Runner','Ghost'],
                      })
movies
```

```python
members = pd.DataFrame({'memid':[1,2,3,4],
                       'name':['Tom','Jon','Aron','Bobby'],
                       'movieid':[1,2,3,np.NaN]
                      })
members
```

```python
# Inner Join By default
pd.merge(movies,members,on='movieid')
```

```python
#Left Join

pd.merge(movies,members,on='movieid',how='left')
```

```python
#Right Join 

pd.merge(movies,members,on='movieid',how='right')
```

```python
# Outer Join 

pd.merge(movies,members,on='movieid',how='outer')
```

```python
# Select only 3 columns from the merged dataframe

pd.merge(movies,members,on='movieid',how='right')[['memid','name','title']]
```

### Handling Duplicates

```python
ddup=DataFrame({'col1':['A','A','B','B','B'],'col2':[2,2,2,3,3]})

ddup
```

```python
# Use duplicated to find duplicate

ddup.duplicated()
```

```python
ddup.drop_duplicates()
```

```python
# You can filter which duplicates to drop by a single columndu.
# ddup_duplicates(['col1')

ddup.drop_duplicates(subset=['col1'])
```

### Group By

```python
dgrp=pd.read_csv('Salaries.csv')
dgrp
```

```python
grp=dgrp.groupby('rank')
```

```python
for rank,group in grp:
    print(rank,group[['discipline','sex','salary']])
```

```python
## GroupBy 2 columns 

grp2=dgrp.groupby(['rank','sex'])
```

```python
for rank,group in grp2:
    print(rank,group[['discipline','sex','salary']])
```

```python
## For 1 particular group 

grp2.get_group(('Prof','Female'))
```
